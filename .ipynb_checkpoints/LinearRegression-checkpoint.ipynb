Linear Regression

```python
# importing necessary libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# reading the file
df = pd.read_csv('HousingData.csv')
```

```python
# removing the rows that have null data
df = df.dropna()
df.isnull().sum()
```

```python
df
```

```python
#linear regression

#initial parameters 
w = np.zeros(13)
b = 0
t = 0 
itr = 1000
lr = 0.01

for t in range(itr):
    ypred = np.dot(df.iloc[:,0:13], w) + b
    error = ypred - df["MEDV"]
    
    dw = 2*np.dot(df.iloc[:, 0:13].T, error)
    db = 2*np.sum(error)

    w -= lr * dw
    b -= lr* db
```
