Linear Regression

```python
# importing necessary libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# reading the file
df = pd.read_csv('HousingData.csv')
```

```python
# removing the rows that have null data
df = df.dropna()
df.isnull().sum()
```

```python
df
```

```python
df.mean()
```

```python
df.std()
```

```python
#linear regression

#initial parameters 
w = np.zeros(13)
b = 0
t = 0 
itr = 10000
lr = 0.01
nrows = df.shape[0]
ncols = df.shape[1] - 1
x = df.iloc[:,0:13]
y = df["MEDV"]

origmean = x.mean()
origstd = x.std()
x = (x-x.mean())/x.std()

for t in range(itr):
    ypred = np.dot(x, w) + b
    error = ypred - df["MEDV"]
    mse = np.sum(error**2)/nrows
    
    wchange = (2/nrows)*np.dot(x.T, error)
    bchange = (2/nrows)*np.sum(error)

    w -= lr * wchange
    b -= lr* bchange

b = b - np.dot(origmean/origstd,w)
w = w/origstd

print("Weight:\n",w)
print("Bias: ",b)
```
