## Implementing logistic regression

```python
X = df.drop(['Survived'], axis=1)
Y = df['Survived']
lr = 0.001
itr = 10000

w = np.zeros(7)
b = w[0]

X = (X-X.mean())/X.std()
origmean = X.mean()
origstd = X.std()

def sigmoid(z):
    z = np.clip(z, -500, 500) #at 500 sigmoid almost equal to 1 and at -500 almost zero
    return 1 / (1 + np.exp(-z))

for i in range(itr):
    z = np.dot(X, w) + b
    yprob = sigmoid(z)
    
    dw = -np.dot(X.T,(Y-yprob)) / X.shape[0]
    db = -np.sum(Y-yprob) / X.shape[0]
    
    w -= lr*dw
    b -= lr*db

b = b - np.dot(origmean/origstd,w)
w = w/origstd

print("Weight:\n",w)
print("Bias: ",b)
```
