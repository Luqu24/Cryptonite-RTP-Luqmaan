```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv')
df.drop(['PassengerId', 'Cabin', 'Name', 'Ticket'], axis=1,inplace=True)
df = df.dropna(subset=['Embarked'])
df.fillna({'Age': df['Age'].mean()}, inplace=True)
df.replace({
    'Sex': {
        'male': 1,
        'female': 0
    },
    'Embarked': {
        'S': 1,
        'Q': 2,
        'C': 3
    }
}, inplace=True)
```

```python
X = df.drop(['Survived'], axis=1)
Y = df['Survived']
log_reg = LogisticRegression(max_iter=10000, penalty=None,solver='lbfgs') 
log_reg.fit(X, Y)
print("Logistic Regression Weights:\n", log_reg.coef_[0])
print("Logistic Regression Bias:", log_reg.intercept_[0])
```
